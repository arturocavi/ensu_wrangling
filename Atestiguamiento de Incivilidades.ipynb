## Atestiguamiento de Incivilidades en los Alrededores de su Vivienda V2
Pregunta 1.4 del cuestionario principal de ENSU: En los últimos tres meses, ¿ha escuchado o ha visto en los alrededores de su vivienda situaciones como...<br>
* BP1_4_1	 vandalismo (grafitis, pintas, vidrios quebrados o daños) en las viviendas o negocios
* BP1_4_2	 consumo de alcohol en las calles
* BP1_4_3	 robos o asaltos
* BP1_4_4	 bandas violentas o pandillerismo
* BP1_4_5	 venta o consumo de droga
* BP1_4_6	 disparos frecuentes con armas
* BP1_4_7	 robo o venta ilegal de gasolina o diésel (huachicol)
* BP1_4_8	 tomas irregulares de luz (energía eléctrica) o diablitos


### Inicio

```python
import pandas as pd
```

#### Importar bases

**Percepción de Inseguridad en la lugares varios** <br>
Fuente: *Encuesta Nacional de Seguridad Pública Urbana 2023 (I Trimestre).* Datos abiertos/2023/Encuesta Nacional de Seguridad Pública Urbana. Trimestre I<br>
Cuestionario principal de la encuesta sección I, II, III, IV (conjunto_de_datos_CB_ENSU_03_2023.csv) <br>
Pregunta *1.4 En los últimos tres meses, ¿ha escuchado o ha visto en los alrededores de su vivienda situaciones como...*<br>
https://www.inegi.org.mx/programas/ensu/#Datos_abiertos

```python
# Importar base de Cuestionario principal de la encuesta sección I, II, III, IV
ensu = pd.read_csv('conjunto_de_datos_CB_ENSU_03_2023.csv')
ensu.replace('\r','',inplace=True)
ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD'.split()] = ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD'.split()].apply(lambda x:x.str.replace('\r',''), axis=1)

#Convertir identificadores de Entidad y Municipio en Texto
ensu['CVE_ENT'] = ensu['CVE_ENT'].astype('str').apply(lambda x:x.zfill(2))
ensu['CVE_MUN'] = ensu['CVE_MUN'].astype('str').apply(lambda x:x.zfill(3))
ensu['ENT_MUN'] = ensu['CVE_ENT'] + ensu['CVE_MUN']
ensu.head()
```

### Filtrar Datos de Interés de ENSU
Diccionario de datos (Metadatos): <br>
https://www.inegi.org.mx/rnm/index.php/catalog/859/data-dictionary/F5?file_name=ENSU_CB_0323 <br>

Opciones de Atestiguamiento para respuestas de variables:
* 1 -> Sí
* 2 -> No
* 9 -> 	No sabe / no responde


```python
#Filtrar columnas de interés de ENSU
ci = 'CD NOM_CD NOM_ENT	BP1_4_1	BP1_4_2	BP1_4_3	BP1_4_4	BP1_4_5	BP1_4_6	BP1_4_7	BP1_4_8	FAC_SEL'.split()
ensuci = ensu[ci]

ensuci['NOM_CD'] = ensuci['NOM_CD'].apply(lambda x:x.title())

#Abreviaturas de Entidades Federativas
ef_abr = {'AGUASCALIENTES':'AGS', 'BAJA CALIFORNIA':'BC', 'BAJA CALIFORNIA SUR':'BCS',
       'CAMPECHE':'CAMP', 'COAHUILA DE ZARAGOZA':'COAH', 'COLIMA':'COL', 'CHIAPAS':'CHIS',
       'CHIHUAHUA':'CHIH', 'CIUDAD DE MEXICO':'CDMX', 'DURANGO':'DGO', 'GUANAJUATO':'GTO',
       'GUERRERO':'GRO', 'HIDALGO':'HGO', 'JALISCO':'JAL', 'MEXICO':'MEX', 'MICHOACAN DE OCAMPO':'MICH',
       'MORELOS':'MOR', 'NAYARIT':'NAY', 'NUEVO LEON':'NL', 'OAXACA':'OAX', 'PUEBLA':'PUE',
       'QUERETARO':'QRO', 'QUINTANA ROO':'QROO', 'SAN LUIS POTOSI':'SLP', 'SINALOA':'SIN',
       'SONORA':'SON', 'TABASCO':'TAB', 'TAMAULIPAS':'TAMPS', 'TLAXCALA':'TLAX',
       'VERACRUZ DE IGNACIO DE LA LLAVE':'VER', 'YUCATAN':'YUC', 'ZACATECAS':'ZAC'}
for ef, abr in ef_abr.items():
    ensuci['NOM_ENT'] = ensuci['NOM_ENT'].replace(ef, abr)

ensuci.loc[ensuci['NOM_CD']=='La Laguna','NOM_ENT'] = 'COAH-DGO'
ensuci.loc[ensuci['NOM_CD']=='Tampico','NOM_ENT'] = 'TAMPS-VER'

ensuci.head()
```

**Ciudades de Interés** <br>
Ciudades con mayor población (de 18 años o más) de acuerdo con dicho levantamiento.

```python
#Ciudades con mayor población
#Población por Ciudad
ppc = ensuci.groupby(['CD','NOM_CD','NOM_ENT'])['FAC_SEL'].sum().reset_index()
ppc.rename(columns = {'FAC_SEL':'POB'}, inplace=True)
# Crear nombre de ciudad, entidad (NOM_CD_ENT)
ppc['NOM_CD_ENT'] = ppc['NOM_CD'] + ', ' + ppc['NOM_ENT']

#Las 20 Ciudades con mayor población
ciu=ppc[['CD','NOM_CD_ENT','POB']].sort_values(by='POB', ascending=False).head(20)
ciu.drop('POB', axis=1, inplace=True)
ciu
```

```python
#Filtrar Ciudades de interés (para trabajar con base de 20 municipios con mayor población)
ensuci = ensuci[ensuci['CD'].isin(ciu['CD'])]
ensuci.drop(['NOM_CD', 'NOM_ENT'],axis=1,inplace=True)
ensuci.info()
```

### Obtención de Porcentaje de Atestiguamiento de Incivilidades para Todas las Variables

```python
# Base inicial de variables
datos_iniciales = ensuci['CD'].unique()
variables = pd.DataFrame(datos_iniciales,columns=['CD'])

# Nombres de variables
nomvar=ensuci.columns[1:len(ensuci.columns)-1]

#Población por Ciudad
ppc = ensuci.groupby('CD')['FAC_SEL'].sum().reset_index()
ppc.rename(columns={'FAC_SEL':'POB'}, inplace=True)

# Loop
for var in nomvar:
    # Atestiguamiento por municipio
    # 1 -> Sí
    apm = ensuci[ensuci[var]==1].groupby('CD')['FAC_SEL'].sum().reset_index()
    apm.rename(columns={'FAC_SEL':'SI_ATE'},inplace=True)
    
    #Merge Población, Sí Atestiguó
    union = ppc.merge(apm, how='left', on='CD')
    union.fillna(0,inplace=True)
    
    #Porcentaje que atestiguó
    union[var]=union['SI_ATE']/union['POB']
    
    #Variable de interés (porcentaje que atestiguó) en Data Frame
    var_df=union[['CD',var]]
    
    #Unión de base inicial con variable
    variables=variables.merge(var_df, how='left',on='CD')
    
variables.head()
```

### Unir Base de Atestiguamiento con Nombres de Ciudades

```python
#Base de percepción de ciudades
atesti = ciu.merge(variables, how='left', on='CD')
atesti.head()
```

#### Cifras Nacionales

```python
#Función para el porcentaje de atestiguamiento nacional
def pct_ate_nac(pob,var):
    nac_ate = ensu[ensu[var]==1]['FAC_SEL'].sum()
    return nac_ate/(pob)

# Nombres de variables
nomvar = atesti.columns[2:len(atesti.columns)]

# Total población de la encuesta
pob = ensu['FAC_SEL'].sum()

# Porcentajes Nacionales con función
nac_pcts = list(map(lambda var: pct_ate_nac(pob,var),nomvar))

atesti.loc[20] = [0, 'Nacional'] + nac_pcts
atesti
```
