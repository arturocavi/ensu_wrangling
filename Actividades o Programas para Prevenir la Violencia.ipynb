### Inicio

```python
import pandas as pd
```

#### Importar Bases

**Percepción de Inseguridad en la lugares varios** <br>
Fuente: *Encuesta Nacional de Seguridad Pública Urbana 2023 (I Trimestre).* Datos abiertos/2023/Encuesta Nacional de Seguridad Pública Urbana. Trimestre I<br>
Cuestionario principal de la encuesta sección I, II, III, IV (conjunto_de_datos_CB_ENSU_03_2023.csv) <br>
Pregunta *3.2a ¿Conoce o ha escuchado si el gobierno de (CIUDAD) ha implementado actividades o programas para prevenir la violencia y/o la delincuencia, tales como, policía de barrio, eventos deportivos, culturales, sociales, jornadas de reforestación, entre otros?*<br>
https://www.inegi.org.mx/programas/ensu/#Datos_abiertos

```python
# Importar base de Cuestionario principal de la encuesta sección I, II, III, IV
ensu = pd.read_csv('conjunto_de_datos_CB_ENSU_03_2023.csv')
ensu.replace('\r','',inplace=True)
ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD BP1_8_2'.split()] = ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD BP1_8_2'.split()].apply(lambda x:x.str.replace('\r',''), axis=1)

#Convertir identificadores de Entidad y Municipio en Texto
ensu['CVE_ENT'] = ensu['CVE_ENT'].astype('str').apply(lambda x:x.zfill(2))
ensu['CVE_MUN'] = ensu['CVE_MUN'].astype('str').apply(lambda x:x.zfill(3))
ensu['ENT_MUN'] = ensu['CVE_ENT'] + ensu['CVE_MUN']
ensu.head()
```

### Filtrar Datos de Interés de ENSU
Diccionario de datos (Metadatos): <br>
https://www.inegi.org.mx/rnm/index.php/catalog/859/data-dictionary/F5?file_name=ENSU_CB_0323 <br>

Primero se tienen que filtrar a los que sí identifican a la Policía Estatal <br>
* BP3_2A	Conocimiento de que el gobierno ha implementado actividades o programas para prevenir la violencia y/o la delincuencia

```python
#Filtrar columnas de interés de ENSU
ci = 'CD NOM_CD NOM_ENT BP3_2A FAC_SEL'.split()
ensuci = ensu[ci]

ensuci['NOM_CD'] = ensuci['NOM_CD'].apply(lambda x:x.title())

#Abreviaturas de Entidades Federativas
ef_abr = {'AGUASCALIENTES':'AGS', 'BAJA CALIFORNIA':'BC', 'BAJA CALIFORNIA SUR':'BCS',
       'CAMPECHE':'CAMP', 'COAHUILA DE ZARAGOZA':'COAH', 'COLIMA':'COL', 'CHIAPAS':'CHIS',
       'CHIHUAHUA':'CHIH', 'CIUDAD DE MEXICO':'CDMX', 'DURANGO':'DGO', 'GUANAJUATO':'GTO',
       'GUERRERO':'GRO', 'HIDALGO':'HGO', 'JALISCO':'JAL', 'MEXICO':'MEX', 'MICHOACAN DE OCAMPO':'MICH',
       'MORELOS':'MOR', 'NAYARIT':'NAY', 'NUEVO LEON':'NL', 'OAXACA':'OAX', 'PUEBLA':'PUE',
       'QUERETARO':'QRO', 'QUINTANA ROO':'QROO', 'SAN LUIS POTOSI':'SLP', 'SINALOA':'SIN',
       'SONORA':'SON', 'TABASCO':'TAB', 'TAMAULIPAS':'TAMPS', 'TLAXCALA':'TLAX',
       'VERACRUZ DE IGNACIO DE LA LLAVE':'VER', 'YUCATAN':'YUC', 'ZACATECAS':'ZAC'}
for ef, abr in ef_abr.items():
    ensuci['NOM_ENT'] = ensuci['NOM_ENT'].replace(ef, abr)

ensuci.loc[ensuci['NOM_CD']=='La Laguna','NOM_ENT'] = 'COAH-DGO'
ensuci.loc[ensuci['NOM_CD']=='Tampico','NOM_ENT'] = 'TAMPS-VER'

ensuci.head()
```

```python
ensuci.info()
```

**Ciudades de Interés** <br>
Ciudades con mayor población (de 18 años o más) de acuerdo con dicho levantamiento.

```python
#Ciudades con mayor población
#Población por Ciudad
ppc = ensuci.groupby(['CD','NOM_CD','NOM_ENT'])['FAC_SEL'].sum().reset_index()
ppc.rename(columns = {'FAC_SEL':'POB'}, inplace=True)
# Crear nombre de ciudad, entidad (NOM_CD_ENT)
ppc['NOM_CD_ENT'] = ppc['NOM_CD'] + ', ' + ppc['NOM_ENT']

#Las 20 Ciudades con mayor población
ciu=ppc[['CD','NOM_CD_ENT','POB']].sort_values(by='POB', ascending=False).head(20)
ciu.drop('POB', axis=1, inplace=True)
ciu
```

```python
#Filtrar Ciudades de interés (para trabajar con base de 20 municipios con mayor población)
ensuci = ensuci[ensuci['CD'].isin(ciu['CD'])]
ensuci.drop(['NOM_CD', 'NOM_ENT'],axis=1,inplace=True)
ensuci.info()
```

### Obtención de Porcentaje de Efectividad del Desempeño de Policías

Opciones de respuestas para pregunta *3.2a ¿Conoce o ha escuchado si el gobierno de (CIUDAD) ha implementado actividades o programas para prevenir la violencia y/o la delincuencia, tales como, policía de barrio, eventos deportivos, culturales, sociales, jornadas de reforestación, entre otros?* 
* 1 -> Sí
* 2 -> No
* 9 -> No sabe / no responde

#### Colapsar datos por municipio

```python
#Población por Municipio
ppm = ensuci.groupby('CD')['FAC_SEL'].sum().reset_index()
ppm.rename(columns = {'FAC_SEL':'POB'}, inplace=True)

#Población que conoce que el gobierno ha implementado actividades o programas para prevenir la violencia y/o la delincuencia
# 1 -> Sí
pcg = ensuci[ensuci['BP3_2A']==1].groupby('CD').sum().reset_index()
pcg.rename(columns={'FAC_SEL':'CON_ACT'}, inplace=True)

#Unión de variables
union = ppm.merge(pcg, how='left', on='CD')

#Porcentaje de conocimiento de actividades del gobierno para prevenir la violencia
union['PCT_CON'] = union['CON_ACT']/union['POB']

#Variable de interés (porcentaje que conoce actividades) en Data Frame
var_df=union[['CD','PCT_CON']]

var_df.head()
```

### Unir Base de Conocimiento con Nombres de Ciudades

```python
#Base de Conocimiento por Ciudades
con_act = ciu.merge(var_df, how='left', on='CD')
con_act
```

#### Cifras Nacionales

```python
# Porcentaje Nacional
nac_pob = ensu['FAC_SEL'].sum() #total población de la encuesta
nac_con = ensu[ensu['BP3_2A'] == 1]['FAC_SEL'].sum() #total población que conoce actividades
nac_con_pct = nac_con/nac_pob

con_act.loc[20] = [0, 'Nacional'] + [nac_con_pct]
con_act
```

```python
#Organizar Tabla a Exportar a Excel
con_act = con_act.sort_values(by='PCT_CON')
con_act
```

#### Tabla 1: Porcentaje de la población de 18 años que mencionó conocer o haber escuchado que el gobierno de la ciudad ha implementado actividades o programas para prevenir la violencia y/o delincuencia de las 20 ciudades con más población
