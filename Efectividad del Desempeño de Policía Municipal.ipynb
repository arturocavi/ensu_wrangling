## Obtención de Efectividad del Desempeño de Policía Municipal V2

Pregunta 1.8 del cuestionario principal de la ENSU: ¿Qué tan efectivo considera el desempeño de la Policía Preventiva Municipal?

### Inicio

```python
import pandas as pd
```

#### Importar Bases

**Percepción de Inseguridad en la lugares varios** <br>
Fuente: *Encuesta Nacional de Seguridad Pública Urbana 2023 (I Trimestre).* Datos abiertos/2023/Encuesta Nacional de Seguridad Pública Urbana. Trimestre I<br>
Cuestionario principal de la encuesta sección I, II, III, IV (conjunto_de_datos_CB_ENSU_03_2023.csv) <br>
Pregunta *1.8 ¿Qué tan efectivo considera el desempeño de la Policía Preventiva Municipal?*<br>
https://www.inegi.org.mx/programas/ensu/#Datos_abiertos

```python
# Importar base de Cuestionario principal de la encuesta sección I, II, III, IV
ensu = pd.read_csv('conjunto_de_datos_CB_ENSU_03_2023.csv')
ensu.replace('\r','',inplace=True)
ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD BP1_8_1'.split()] = ensu['ID_PER NOM_ENT NOM_MUN NOM_CD TIPO_CUEST DOMINIO REG_CD BP1_8_1'.split()].apply(lambda x:x.str.replace('\r',''), axis=1)

#Convertir identificadores de Entidad y Municipio en Texto
ensu['CVE_ENT'] = ensu['CVE_ENT'].astype('str').apply(lambda x:x.zfill(2))
ensu['CVE_MUN'] = ensu['CVE_MUN'].astype('str').apply(lambda x:x.zfill(3))
ensu['ENT_MUN'] = ensu['CVE_ENT'] + ensu['CVE_MUN']
ensu.head()
```

### Filtrar Datos de Interés de ENSU
Diccionario de datos (Metadatos): <br>
https://www.inegi.org.mx/rnm/index.php/catalog/859/data-dictionary/F5?file_name=ENSU_CB_0323 <br>

Primero se tienen que filtrar a los que sí identifican a la Policía Preventiva Municipal <br>
* BP1_7_1	Identificación de la Policía Preventiva Municipal
* BP1_8_1	Percepción de la efectividad del desempeño de la Policía Preventiva Municipal

```python
#Filtrar columnas de interés de ENSU
ci = 'CD NOM_CD NOM_ENT BP1_7_1 BP1_8_1 FAC_SEL'.split()
ensuci = ensu[ci]

ensuci['NOM_CD'] = ensuci['NOM_CD'].apply(lambda x:x.title())

#Abreviaturas de Entidades Federativas
ef_abr = {'AGUASCALIENTES':'AGS', 'BAJA CALIFORNIA':'BC', 'BAJA CALIFORNIA SUR':'BCS',
       'CAMPECHE':'CAMP', 'COAHUILA DE ZARAGOZA':'COAH', 'COLIMA':'COL', 'CHIAPAS':'CHIS',
       'CHIHUAHUA':'CHIH', 'CIUDAD DE MEXICO':'CDMX', 'DURANGO':'DGO', 'GUANAJUATO':'GTO',
       'GUERRERO':'GRO', 'HIDALGO':'HGO', 'JALISCO':'JAL', 'MEXICO':'MEX', 'MICHOACAN DE OCAMPO':'MICH',
       'MORELOS':'MOR', 'NAYARIT':'NAY', 'NUEVO LEON':'NL', 'OAXACA':'OAX', 'PUEBLA':'PUE',
       'QUERETARO':'QRO', 'QUINTANA ROO':'QROO', 'SAN LUIS POTOSI':'SLP', 'SINALOA':'SIN',
       'SONORA':'SON', 'TABASCO':'TAB', 'TAMAULIPAS':'TAMPS', 'TLAXCALA':'TLAX',
       'VERACRUZ DE IGNACIO DE LA LLAVE':'VER', 'YUCATAN':'YUC', 'ZACATECAS':'ZAC'}
for ef, abr in ef_abr.items():
    ensuci['NOM_ENT'] = ensuci['NOM_ENT'].replace(ef, abr)

ensuci.loc[ensuci['NOM_CD']=='La Laguna','NOM_ENT'] = 'COAH-DGO'
ensuci.loc[ensuci['NOM_CD']=='Tampico','NOM_ENT'] = 'TAMPS-VER'

ensuci.head()
```

**Ciudades de Interés** <br>
Ciudades con mayor población (de 18 años o más) de acuerdo con dicho levantamiento excluyendo a CDMX porque no aplica.

```python
#Ciudades con mayor población
#Población por Ciudad
#Filtrar municipios de interés (para trabajar con base de 20 municipios con mayor población excluyendo a CDMX porque no aplica)
ppc = ensuci[ensuci['NOM_ENT']!='CDMX'].groupby(['CD','NOM_CD','NOM_ENT'])['FAC_SEL'].sum().reset_index()
ppc.rename(columns = {'FAC_SEL':'POB'}, inplace=True)
# Crear nombre de ciudad, entidad (NOM_CD_ENT)
ppc['NOM_CD_ENT'] = ppc['NOM_CD'] + ', ' + ppc['NOM_ENT']

#Las 20 Ciudades con mayor población
ciu=ppc[['CD','NOM_CD_ENT','POB']].sort_values(by='POB', ascending=False).head(20)
ciu.drop('POB', axis=1, inplace=True)
ciu
```

```python
#Filtrar Ciudades de interés (para trabajar con base de 20 municipios con mayor población)
ensuci = ensuci[ensuci['CD'].isin(ciu['CD'])]
ensuci.drop(['NOM_CD', 'NOM_ENT'],axis=1,inplace=True)
ensuci.info()
```

#### Filtrar personas que identifican a la polcía preventiva municipal
BP1_8_1 Percepción de la efectividad del desempeño de la Policía Preventiva Municipal <br>
Opciones de respuestas para pregunta 1.7 *¿Usted identifica a la Policía Preventiva Municipal?* <br>
* Sí -> 1
* No -> 2
* No aplica -> 3
* No sabe / no responde -> 9

```python
# Filtrar personas que identifican a la polcía preventiva municipal
ensuci = ensuci[ensuci['BP1_7_1'] == 1]
ensuci.drop('BP1_7_1', axis=1, inplace=True)

#Convertir variable a numérica
ensuci['BP1_8_1'] = ensuci['BP1_8_1'].astype('int')

ensuci.info()
```

### Obtención de Porcentaje de Efectividad del Desempeño de Policías

Opciones de respuestas para pregunta *1.8 ¿Qué tan efectivo considera el desempeño de la Policía Preventiva Municipal?* 
* 1 -> Muy efectivo
* 2 -> Algo efectivo
* 3 -> Poco efectivo
* 4 -> Nada efectivo
* 9 -> 	No sabe / no responde

#### Colapsar datos por municipio

```python
#Población por Municipio
ppm = ensuci.groupby('CD')['FAC_SEL'].sum().reset_index()
ppm.rename(columns = {'FAC_SEL':'POB'}, inplace=True)

#Población que considera efectiva a la policía
# 1 -> Muy efectivo
# 2 -> Algo efectivo
# 3 -> Poco efectivo
pef = ensuci[(ensuci['BP1_8_1'] == 1) | (ensuci['BP1_8_1'] == 2)]. groupby('CD')['FAC_SEL'].sum().reset_index()
pef.rename(columns = {'FAC_SEL':'EFECTIVA'}, inplace=True)

#Unión de variables
union = ppm.merge(pef, how='left', on='CD')

#Porcentaje de efectividad de policías
union['PCT_EFE'] = union['EFECTIVA']/union['POB']

#Variable de interés (porcentaje que se siente inseguro) en Data Frame
var_df=union[['CD','PCT_EFE']]

var_df.head()
```

### Unir Base de Efectividad de Policías con Nombres de Estados y Municipios

```python
#Base de Efectividad por Municipios
efec_poli = ciu.merge(var_df, how='left', on='CD')
efec_poli
```

#### Cifras Nacionales

```python
# Porcentaje Nacional
nac_pob = ensu['FAC_SEL'].sum() #total población de la encuesta
nac_pob_iap = ensu[ensu['BP1_7_1'] == 1]['FAC_SEL'].sum() #total población de la encuesta que identifica a la policía
nac_efe = ensu[(ensu['BP1_8_1'] == '1') | (ensu['BP1_8_1'] == '2')]['FAC_SEL'].sum()
nac_efe_pct = nac_efe/nac_pob_iap

efec_poli.loc[20] = [0, 'Nacional'] + [nac_efe_pct]
efec_poli
```

```python
#Organizar Tabla a Exportar a Excel
efec_poli = efec_poli.sort_values(by='PCT_EFE')
efec_poli
```

#### Tabla 1: Porcentaje de la población de 18 años que consideró efectivo el desempeño de la Policía Preventiva Municipal de las 20 ciudades con más población
